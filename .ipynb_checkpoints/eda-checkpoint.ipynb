# Global design

```python
plt.rcParams.update({
    'axes.titlesize': 20,
    'axes.titlepad':20,
    'axes.titleweight': 'bold',
    'axes.labelsize': 16,
    'axes.labelpad': 20,
    'axes.labelweight': 'bold',
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
    'figure.figsize': [10,6]
})

def custom_bar_params():
    return {
            'palette' : 'Blues_r',
            'saturation' : 1,
            'edgecolor' : 'lightgrey',
            'width' : .7,
            'legend' : False
    }

def custom_pie_params():
    return {
            'color' : 'white',
            'fontsize' : 12,
            'fontweight' : 'bold'
    }
```

## Remote job percentage for all job titles

```python
#custom function to get remote job percentage for all job titles
def remote_comparison(job_title, ax):
    df_remote = df[df['job_title_short'] == job_title]['job_work_from_home'].value_counts()
    df_remote = df_remote.rename({0 : 'On-site', 1: 'Remote'})

    ax.pie(
            df_remote,
            labels=df_remote.index,
            autopct=lambda p: f'{p:.1f}%',
            startangle=20,
            explode=[.05, 0.05],
            textprops=custom_pie_params()
    )
    ax.set_title(f'Percentage of remote {job_title} jobs', fontsize=16, pad=20, fontweight='bold')


# Unique job titles
job_titles = df['job_title_short'].unique()

# Create subplots
fig, axes = plt.subplots(2, 3, figsize=(20,16))

axes = axes.flatten()

for ax, job_title in zip(axes, job_titles):
    remote_comparison(job_title, ax)


axes[0].legend(fontsize = 12)
plt.tight_layout()
plt.show()

```

# Monthly job trends

```python
months_order = [
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
]

job_trend_month = (
                    df['job_posted_date']
                    .apply(lambda x : x.strftime('%B'))
                    .value_counts()
                    )

df_job_trend_month  = pd.DataFrame(job_trend_month).reset_index()
df_job_trend_month.columns = ['Job posted month', 'Number of jobs']

df_job_trend_month['Job posted month'] = pd.Categorical(
                                                        df_job_trend_month['Job posted month'],
                                                        categories=months_order,
                                                        ordered=True
                                                        )

df_job_trend_month = df_job_trend_month.sort_values(by='Job posted month')
df_job_trend_month

```

```python
sns.lineplot(df_job_trend_month, x='Job posted month', y='Number of jobs')
plt.title('Monthly trend of data jobs')
plt.xticks(rotation = 45)

plt.xlabel('Month')
plt.ylabel('Number of total jobs')
plt.tight_layout()
plt.show()



```

# Skill trend throughout the year

```python
df_skill_trend = df[['job_posted_date', 'job_skills']].copy()
df_skill_trend['job_posted_date'] = pd.to_datetime(df_skill_trend['job_posted_date'])
df_skill_trend['month'] = df_skill_trend['job_posted_date'].dt.month
df_skill_trend = df_skill_trend.drop(columns='job_posted_date').explode('job_skills')
df_skill_trend_pivot = df_skill_trend.pivot_table(index='month',columns='job_skills', aggfunc='size',fill_value=0)
df_skill_trend_pivot.loc['total'] = df_skill_trend_pivot.sum()
top_5_skills_trend = pd.DataFrame(df_skill_trend_pivot[df_skill_trend_pivot.loc['total'].sort_values(ascending = False).index].drop('total').iloc[:,:5])

import calendar
top_5_skills_trend.index = top_5_skills_trend.index.map(lambda x: calendar.month_name[x])
top_5_skills_trend.index = pd.CategoricalIndex(top_5_skills_trend.index, categories=months_order, ordered=True)
top_5_skills_trend.sort_index()
top_5_skills_trend
```

```python
sns.lineplot(top_5_skills_trend, dashes=False, linewidth = 1)

plt.title('Monthly trend of skills')
plt.xlabel('')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Average and median salary by job titles

```python
mean_salary = (
              df.groupby(by='job_title_short')['salary_year_avg']
              .mean()
              .apply(lambda x : int(x))
              .sort_values(ascending = False)
              )

median_salary = (
                df.groupby(by='job_title_short')['salary_year_avg']
                .median()
                .apply(lambda x : int(x))
                .sort_values(ascending = False)
                )

salary_df = pd.DataFrame({
    'job_title' : mean_salary.index,
    'mean_salary' : mean_salary.values,
    'median_salary' : median_salary.values
})

salary_df_melted = salary_df.melt(id_vars='job_title', var_name='Type', value_name='Salary')

salary_df
```

```python
sns.barplot(
            salary_df_melted,
            x='Salary',
            y='job_title',
            hue='Type'
            )

plt.title('Mean and median salary by job title')
plt.xlabel('Salary')
plt.ylabel('')
custom_handles = [mpatches.Patch(color=sns.color_palette()[0], label='Mean salary'),          #custom legend
                mpatches.Patch(color=sns.color_palette()[1], label='Median salary')]
plt.legend(handles = custom_handles)
plt.xlim(left=70000)
plt.tight_layout()
plt.show()
```

# Top country according to salary 

```python
df_salary_country = df[['job_country', 'salary_year_avg']].dropna(subset ='salary_year_avg')

country_counts = df_salary_country['job_country'].value_counts()
countries_withatleast_50_values = country_counts[country_counts.values > 50].index   # at least 50 salary value check

filtered_country = df_salary_country[df_salary_country['job_country'].isin(countries_withatleast_50_values)]
df_median_salary_country = pd.DataFrame(
                            filtered_country
                            .groupby('job_country')['salary_year_avg']
                            .median()
                            .sort_values(ascending = False)
                            .head(10)
                            ).reset_index()

df_median_salary_country.columns = ['Country', "Average yearly salary"]
df_median_salary_country
```

```python
sns.barplot( 
            df_median_salary_country,
            x='Average yearly salary',
            y='Country',
            hue='Country',
            **custom_bar_params()
            )

plt.title('Median salary of top 10 country')
plt.xlabel('Yearly salary ($)')
plt.ylabel('')
plt.xlim(left = 80000)
plt.tight_layout()
plt.show()
```
